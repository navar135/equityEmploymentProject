```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./rawData/NationalHealthInterview/2019NHIS_raw.csv")
```

```python
df.head(10)
```

```python
df.columns
```

#### Column - URBRRL

```python
dfnew = df.rename(columns={'URBRRL':'Urban-Rural Classification'})
dfnew['Urban-Rural Classification'].replace({1:"Large central metro", 2:"Large fringe metro", 3:"Medium and small metro",
                                           4:"Nonmetropolitan"}, inplace=True)

#check for null values
print("Number of null rows: ", dfnew['Urban-Rural Classification'].isnull().sum())

#dfnew.groupby('Urban-Rural Classification').count()
x = pd.DataFrame(dfnew['Urban-Rural Classification'].value_counts())
display(x)
x.plot.bar()
```

#### Column - INCGRP_A 

```python
dfnew = dfnew.rename(columns={'INCGRP_A':'Family_Income(USD)'})
dfnew['Family_Income(USD)'].replace({1:"0 to 34,999", 2:"35,000 to 49,999", 3:"50,000 to 74,999",
                                           4:"75,000 to 99,999", 5:"100,000 or greater", 8:"Not Ascertained"}, inplace=True)

#check for null values
print("Number of null rows: ", dfnew['Family_Income(USD)'].isnull().sum())

#dfnew.groupby('Urban-Rural Classification').count()
x = pd.DataFrame(dfnew['Family_Income(USD)'].value_counts())
display(x)
x.plot.bar()
```
